```python
import pandas as pd
data = pd.read_csv('BlackFriday.csv')
```

```python
data['Gender'] = data['Gender'].map({'M':0, 'F':1})
```

```python
df = pd.get_dummies(data['City_Category'])
data = pd.concat([data, df], axis = 1)
data.drop('City_Category', axis = 1, inplace=True)
```

```python
df = pd.get_dummies(data['Age'])
data = pd.concat([data, df], axis = 1)
data.drop('Age', axis = 1, inplace=True)
```

```python
df = pd.get_dummies(data['Occupation'],prefix='Occupation')
data = pd.concat([data, df], axis = 1)
data.drop('Occupation', axis = 1, inplace=True)
```

```python
data.head()
```

```python
df = pd.get_dummies(data['Stay_In_Current_City_Years'],prefix='Stay_In_Current_City_Years')
data = pd.concat([data, df], axis = 1)
data.drop('Stay_In_Current_City_Years', axis = 1, inplace=True)
```

```python
data=data.fillna(0) 
```
